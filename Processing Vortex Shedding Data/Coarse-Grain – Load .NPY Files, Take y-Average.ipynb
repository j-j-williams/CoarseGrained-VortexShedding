```python
###  Loop thru each .NPY file and extract the flowfield data along  between  y_CG_1  <=  y  <=  y_CG_2


U_yPt = np.zeros( ( nx-1 , nt ) )
V_yPt = np.zeros( ( nx-1 , nt ) )
W_yPt = np.zeros( ( nx-1 , nt ) )



for it in range( nt ):
    print('i  =  ' + str(it) + '  /  ' + str(nt-1) )

    U = np.load( stem + 'U/' + 'U_' + str(it) + '.npy' )
    V = np.load( stem + 'V/' + 'V_' + str(it) + '.npy' )
    W = np.load( stem + 'W/' + 'W_' + str(it) + '.npy' )


    
    U_yPt[ : , it ] =    np.mean( U[ : , y_CG_1 : y_CG_2 ] , axis=1 )
    V_yPt[ : , it ] =    np.mean( V[ : , y_CG_1 : y_CG_2 ] , axis=1 )
    W_yPt[ : , it ] =    np.mean( W[ : , y_CG_1 : y_CG_2 ] , axis=1 )
    
    
#     U_yPt[ : , it ] = U[ : , y_CG ]
#     V_yPt[ : , it ] = V[ : , y_CG ]
#     W_yPt[ : , it ] = W[ : , y_CG ]


print('\n\nFinished!!')

```
