```python
###  Load the data


stem = '/Users/josephwilliams/Documents/Data_Results/Coarse-Grained-Vortex-Shedding/Vortex Shedding Data/Coarse-Grained Data/'
stem_CG    = stem + 'y-avg/Re'     + str(Re) + '/'
stem_param = stem + 'Parameters/Re' + str(Re) + '/'



U_cg_n1_0 = np.load(stem_CG + 'U/U_y-1_y0.npy')
U_cg_n2_0 = np.load(stem_CG + 'U/U_y-2_y0.npy')
U_cg_n4_0 = np.load(stem_CG + 'U/U_y-4_y0.npy')
U_cg_n8_0 = np.load(stem_CG + 'U/U_y-8_y0.npy')

U_cg_0_p1 = np.load(stem_CG + 'U/U_y0_y1.npy')
U_cg_0_p2 = np.load(stem_CG + 'U/U_y0_y2.npy')
U_cg_0_p4 = np.load(stem_CG + 'U/U_y0_y4.npy')
U_cg_0_p8 = np.load(stem_CG + 'U/U_y0_y8.npy')



V_cg_n1_0 = np.load(stem_CG + 'V/V_y-1_y0.npy')
V_cg_n2_0 = np.load(stem_CG + 'V/V_y-2_y0.npy')
V_cg_n4_0 = np.load(stem_CG + 'V/V_y-4_y0.npy')
V_cg_n8_0 = np.load(stem_CG + 'V/V_y-8_y0.npy')

V_cg_0_p1 = np.load(stem_CG + 'V/V_y0_y1.npy')
V_cg_0_p2 = np.load(stem_CG + 'V/V_y0_y2.npy')
V_cg_0_p4 = np.load(stem_CG + 'V/V_y0_y4.npy')
V_cg_0_p8 = np.load(stem_CG + 'V/V_y0_y8.npy')



W_cg_n1_0 = np.load(stem_CG + 'W/W_y-1_y0.npy')
W_cg_n2_0 = np.load(stem_CG + 'W/W_y-2_y0.npy')
W_cg_n4_0 = np.load(stem_CG + 'W/W_y-4_y0.npy')
W_cg_n8_0 = np.load(stem_CG + 'W/W_y-8_y0.npy')

W_cg_0_p1 = np.load(stem_CG + 'W/W_y0_y1.npy')
W_cg_0_p2 = np.load(stem_CG + 'W/W_y0_y2.npy')
W_cg_0_p4 = np.load(stem_CG + 'W/W_y0_y4.npy')
W_cg_0_p8 = np.load(stem_CG + 'W/W_y0_y8.npy')




x = np.load(stem_param + 'x.npy')
t = np.load(stem_param + 't.npy')

```

```python
###  Plot the data


# Plot coarse-grained  x-velocity  at specified  y
plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.pcolormesh( x[1:-1] , t , U_cg_n8_0.T , cmap='coolwarm' )   # x[1:-1] , t 
plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained x-Velocity, $-8 \leq y \leq 0$'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.pcolormesh( x[1:-1] , t , U_cg_0_p8.T , cmap='coolwarm' )   # x[1:-1] , t 
plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained x-Velocity, $ 0 \leq y \leq 8$'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.show()


# Plot coarse-grained  y-velocity  at specified  y
plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.pcolormesh( x[1:-1] , t , V_cg_n8_0.T , cmap='coolwarm' )   # x[1:-1] , t 
plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained y-Velocity, $-8 \leq y \leq 0$'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.pcolormesh( x[1:-1] , t , V_cg_0_p8.T , cmap='coolwarm' )   # x[1:-1] , t 
plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained y-Velocity, $ 0 \leq y \leq 8$'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.show()


# Plot coarse-grained  vorticity  at specified  y
plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.pcolormesh( x[1:-1] , t , W_cg_n8_0.T , cmap='coolwarm' )   # x[1:-1] , t 
plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained Vorticity, $-8 \leq y \leq 0$'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.pcolormesh( x[1:-1] , t , W_cg_0_p8.T , cmap='coolwarm' )   # x[1:-1] , t 
plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained Vorticity, $ 0 \leq y \leq 8$'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.show()

```
