```python
###  Load the data


stem = '/Users/josephwilliams/Documents/Data_Results/Coarse-Grained-Vortex-Shedding/Vortex Shedding Data/Coarse-Grained Data/'
stem_CG    = stem + 'y-avg/Re'     + str(Re) + '/'
stem_param = stem + 'Parameters/Re' + str(Re) + '/'


U_cg_1 = np.load(stem_CG + 'U/U_y-1_y1.npy')
U_cg_2 = np.load(stem_CG + 'U/U_y-2_y2.npy')
U_cg_4 = np.load(stem_CG + 'U/U_y-4_y4.npy')
U_cg_8 = np.load(stem_CG + 'U/U_y-8_y8.npy')


V_cg_1 = np.load(stem_CG + 'V/V_y-1_y1.npy')
V_cg_2 = np.load(stem_CG + 'V/V_y-2_y2.npy')
V_cg_4 = np.load(stem_CG + 'V/V_y-4_y4.npy')
V_cg_8 = np.load(stem_CG + 'V/V_y-8_y8.npy')


W_cg_1 = np.load(stem_CG + 'W/W_y-1_y1.npy')
W_cg_2 = np.load(stem_CG + 'W/W_y-2_y2.npy')
W_cg_4 = np.load(stem_CG + 'W/W_y-4_y4.npy')
W_cg_8 = np.load(stem_CG + 'W/W_y-8_y8.npy')


x = np.load(stem_param + 'x.npy')
t = np.load(stem_param + 't.npy')

```
