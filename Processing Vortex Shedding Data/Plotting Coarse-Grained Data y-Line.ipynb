```python
###  Load the data


stem = '/Users/josephwilliams/Documents/Data_Results/Coarse-Grained-Vortex-Shedding/Vortex Shedding Data/Coarse-Grained Data/'
stem_CG    = stem + 'y-line/Re'     + str(Re) + '/'
stem_param = stem + 'Parameters/Re' + str(Re) + '/'


U_cg_n1 = np.load(stem_CG + 'U/U_y-1.npy')
U_cg_n2 = np.load(stem_CG + 'U/U_y-2.npy')
U_cg_n4 = np.load(stem_CG + 'U/U_y-4.npy')
U_cg_n8 = np.load(stem_CG + 'U/U_y-8.npy')
U_cg_0  = np.load(stem_CG + 'U/U_y0.npy')
U_cg_p1 = np.load(stem_CG + 'U/U_y1.npy')
U_cg_p2 = np.load(stem_CG + 'U/U_y2.npy')
U_cg_p4 = np.load(stem_CG + 'U/U_y4.npy')
U_cg_p8 = np.load(stem_CG + 'U/U_y8.npy')


V_cg_n1 = np.load(stem_CG + 'V/V_y-1.npy')
V_cg_n2 = np.load(stem_CG + 'V/V_y-2.npy')
V_cg_n4 = np.load(stem_CG + 'V/V_y-4.npy')
v_cg_n8 = np.load(stem_CG + 'V/V_y-8.npy')
V_cg_0  = np.load(stem_CG + 'V/V_y0.npy')
V_cg_p1 = np.load(stem_CG + 'V/V_y1.npy')
V_cg_p2 = np.load(stem_CG + 'V/V_y2.npy')
V_cg_p4 = np.load(stem_CG + 'V/V_y4.npy')
v_cg_p8 = np.load(stem_CG + 'V/V_y8.npy')


W_cg_n1 = np.load(stem_CG + 'W/W_y-1.npy')
W_cg_n2 = np.load(stem_CG + 'W/W_y-2.npy')
W_cg_n4 = np.load(stem_CG + 'W/W_y-4.npy')
W_cg_n8 = np.load(stem_CG + 'W/W_y-8.npy')
W_cg_0  = np.load(stem_CG + 'W/W_y0.npy')
W_cg_p1 = np.load(stem_CG + 'W/W_y1.npy')
W_cg_p2 = np.load(stem_CG + 'W/W_y2.npy')
W_cg_p4 = np.load(stem_CG + 'W/W_y4.npy')
W_cg_p8 = np.load(stem_CG + 'W/W_y8.npy')


x = np.load(stem_param + 'x.npy')
t = np.load(stem_param + 't.npy')

```

```python
###  Plot the data


# Plot coarse-grained  x-velocity  at specified  y
plt.figure()
plt.pcolormesh( x[1:-1] , t , U_cg_0.T , cmap='coolwarm' )   # x[1:-1] , t 

plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained x-Velocity'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.show()


# Plot coarse-grained  y-velocity  at specified  y
plt.figure()
plt.pcolormesh( x[1:-1] , t , V_cg_0.T , cmap='coolwarm' )   # x[1:-1] , t 

plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained y-Velocity'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.show()


# Plot coarse-grained  vorticity  at specified  y
plt.figure()
plt.pcolormesh( x[1:-1] , t , W_cg_0.T , cmap='coolwarm' )   # x[1:-1] , t 

plt.xlabel('x' , fontsize=16)
plt.ylabel('t' , fontsize=16)
ttl_str = 'Coarse-Grained Vorticity'
plt.title(ttl_str , fontsize=16)
plt.colorbar()

plt.show()

```

```python
###  Plot slices of the coarse-grained data


t_plt = 0


# Plot coarse-grained  x-velocity  at specified  y
plt.figure()

plt.plot( x[1:-1] , U_cg_0[:,t_plt] )
plt.xlabel('x' , fontsize=16)
plt.ylabel('$\overline{u}$' , fontsize=16)
ttl_str = 'Coarse-Grained x-Velocity at y=0'
plt.title(ttl_str , fontsize=16)

plt.show()


# Plot coarse-grained  y-velocity  at specified  y
plt.figure()

plt.plot( x[1:-1] , V_cg_0[:,t_plt] )
plt.xlabel('x' , fontsize=16)
plt.ylabel('$\overline{v}$' , fontsize=16)
ttl_str = 'Coarse-Grained y-Velocity at y=0'
plt.title(ttl_str , fontsize=16)

plt.show()


# Plot coarse-grained  vorticity  at specified  y
plt.figure()

plt.plot( x[1:-1] , W_cg_0[:,t_plt] )
plt.xlabel('x' , fontsize=16)
plt.ylabel('$\overline{\omega}$' , fontsize=16)
ttl_str = 'Coarse-Grained Vorticity at y=0'
plt.title(ttl_str , fontsize=16)

plt.show()

```
